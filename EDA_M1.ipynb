### The intel processors are of 4 types of series

1. **U Series**:

    - Designed for **ultra-low power consumption**.
    - Commonly used in **thin-and-light laptops** and **ultrabooks**.
    - Optimized for longer battery life, making them ideal for portability.
2. **H Series**:

    - Focuses on **high performance**, often with **higher power consumption**.
    - Popular in **gaming laptops** and **workstations**.
    - Offers excellent performance for intensive tasks like video editing or 3D rendering.
3. **HQ Series** (historically):

    - Similar to the H series, with **quad-core processors** and high performance, Gaming
    - The HQ designation was used in earlier generations (e.g., 6th, 7th Gen) but has been largely replaced by H-series processors in recent generations.
4. **HX Series**:

    - **Extreme performance** for high-power workloads.
    - Found in **desktop-replacement laptops** and systems for **heavy gaming** or **professional applications**.
    - Typically have **higher TDP (Thermal Design Power)** and are capable of desktop-grade performance in a portable form factor.

### Handling outliers

```python
# Defining a function for removing outliers
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
# Creating a new datframe saving the valid values alone
df_cleaned = remove_outliers_iqr(df, 'Weight')
df_cleaned = remove_outliers_iqr(df_cleaned, 'Inches')
```

### GPU performance

```python
import re

def classify_gpu(gpu):
    gpu = gpu.lower()
    # Low-End
    if re.search(r'intel hd|intel uhd|intel iris|arm mali|radeon r2|radeon r3|radeon r4|radeon r5|radeon 520|radeon 530|radeon 540|'
                r'radeon r7 graphics|radeon r7 m440|radeon r7 m445|radeon r7 m460|radeon r7 m365x|radeon r7 m465|'
                r'geforce 920|geforce 930|gt 940mx|geforce 940m|geforce 930mx|nvidia geforce 150mx|amd r4 graphics|'
                r'intel graphics 620', gpu):
        return 'Low-End'
    # Mid-Range
    elif re.search(r'mx130|mx150|mx150mx|940mx|gtx 950m|gtx 960m|gtx 965m|gtx 1050|rx 550|rx 540|radeon pro 455|'
                r'radeon pro 555|radeon pro 560|radeon r9 m385|amd r17m-m1-70|gtx 1050 ti|radeon r7', gpu):
        return 'Mid-Range'
    # High-End
    elif re.search(r'gtx 1060|gtx 1070|gtx 1080|rx 580|rx 560|gtx 1050 ti|gtx 970m|gtx 980|nvidia geforce gtx 1080|nvidia geforce 960m', gpu):
        return 'High-End'
    # Workstation
    elif re.search(r'quadro|firepro', gpu):
        return 'Workstation'
    else:
        return 'Unknown'

# Apply classification function
# Ensure that 'Gpu' column exists in df_cleaned
df_cleaned['Performance_Tier_of_GPU'] = df_cleaned['Gpu'].apply(classify_gpu)
```

```python
df_cleaned['Performance_Tier_of_GPU'].value_counts()
```

```python
# Manually map 'Unknown' GPUs to their correct performance tier
gpu_tier_mapping = {
    'Nvidia GeForce GTX 960<U+039C>': 'Mid-Range',
    'Nvidia GeForce GTX 960': 'Mid-Range',
    'Nvidia GeForce GTX1080': 'High-End',
    'Nvidia GeForce GTX1050 Ti': 'Mid-Range',
    'Nvidia GeForce GTX 930MX': 'Low-End'
}
```

```python
# Update 'Performance_Tier_of_GPU' based on the manual mapping
df_cleaned['Performance_Tier_of_GPU'] = df_cleaned.apply(
    lambda row: gpu_tier_mapping.get(row['Gpu'], row['Performance_Tier_of_GPU']), axis=1
)
```
